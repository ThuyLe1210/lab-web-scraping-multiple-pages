## Find other lists of hot songs on the internet and scrape them too: having a bigger pool of songs will be awesome!

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.last.fm/tag/rolling+stone+500+greatest+songs+of+all+time/tracks?page=1"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.prettify()
```

```python
soup.find_all("td", "chartlist-name")
                                        
                                        


```

```python
titles = []
for title in soup.find_all("td", "chartlist-name"):
     titles.append(title.get_text().strip()) 
titles
```

```python
soup.find_all("td", "chartlist-artist")
                                        
                                        


```

```python
artists = []
for artist in soup.find_all("td", "chartlist-artist"):
     artists.append(artist.get_text().strip()) 
artists
```

```python
songs_of_all_time = pd.DataFrame({"title":titles,
                       "artist":artists,
                      })
songs_of_all_time
```

### Practice web scraping

```python
html = 'https://www.fbi.gov/wanted/topten'
```

```python
response = requests.get(html)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print(soup.prettify())
```

```python
soup.select("div.col-lg-12 a")[0]
```

```python
soup.select('h3.title a')
```

```python
name = []

num_iter = len(soup.select("h3.title a"))

namelist = soup.select("h3.title a")
for i in range(num_iter):
    name.append(namelist[i].get_text())

print(name)
```

```python
names = pd.DataFrame({"name":name})
names
```
